```python
#Distribution of depression score
#Frequency = count of people in each score range
plt.figure() 
plt.hist(df["depression_score"], bins=40,)
plt.xlabel("Depression Score")
plt.ylabel("Frequency")
plt.title("Distribution of Depression Scores")
plt.show()
```

```python
#Average depression score by gender
avg_dep = df.groupby("gender")["depression_score"].mean()

plt.figure()
plt.bar(avg_dep.index, avg_dep.values)
plt.xlabel("Gender")
plt.ylabel("Average Depression Score")
plt.title("Average Depression Score by Gender")
plt.tight_layout()
plt.show()
```

```python
#Stress level distribution
plt.figure()
plt.hist(df["stress_level"], bins=20)
plt.xlabel("Stress Level")
plt.ylabel("Count")
plt.title("Stress Level Distribution")
plt.show()
```

```python
#Sleep hours vs depression score
plt.figure()
plt.scatter(df["sleep_hours"], df["depression_score"])
plt.xlabel("Sleep Hours")
plt.ylabel("Depression Score")
plt.title("Sleep Hours vs Depression")
plt.show()
```

```python
#Mental health risk counts
risk_counts = df["mental_health_risk"].value_counts()

plt.figure()
risk_counts.plot(kind="bar")
plt.xlabel("Mental Health Risk")
plt.ylabel("Count")
plt.title("Mental Health Risk Distribution")
plt.show()
```

```python
# Anxiety vs Depression
plt.figure()
plt.scatter(
    df["depression_score"],
    df["anxiety_score"],
    alpha=0.6
)
plt.xlabel("Depression Score")
plt.ylabel("Anxiety Score")
plt.title("Anxiety Score vs Depression Score")
plt.tight_layout()
plt.show()
```

```python
# Stress vs Anxiety
plt.figure()
plt.scatter(
    df["stress_level"],
    df["anxiety_score"],
    alpha=0.6
)
plt.xlabel("Stress Level")
plt.ylabel("Anxiety Score")
plt.title("Stress Level vs Depression Score")
plt.tight_layout()
plt.show()
```

```python
#Productivity vs mental health
plt.figure()
plt.scatter(df["productivity_score"], df["depression_score"])
plt.xlabel("Productivity Score")
plt.ylabel("Depression Score")
plt.title("Productivity vs Depression")
plt.show()
```
